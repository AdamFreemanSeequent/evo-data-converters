# Publish Geoscience objects from an ADAMF file

This notebook shows how you can login to Evo and publish Geoscience objects from an ADAMF file to your chosen workspace.

An ADAMF file is a CSV file containing `X`, `Y` and `Z` point coordinates. Any additional columns are converted into point attributes, and the whole file is published as a Pointset object.

In this first cell we create a ServiceManagerWidget which will open a browser window and ask you to login.

Once logged in, a widget will be displayed below allowing you to select an organisation and workspace to publish objects to.

__Required:__ You must fill in your `client_id` before running this cell.


```python
from evo.notebooks import ServiceManagerWidget

client_id = "<your-client-id>"  # Replace with your client ID
redirect_url = "<your-redirect-url>"  # Replace with your redirect URL

manager = await ServiceManagerWidget.with_auth_code(
    client_id=client_id,
    redirect_url=redirect_url,
).login()
```

In the cell below we choose the ADAMF file we want to publish and set its path in the `adamf_file` variable.

Then we choose an EPSG code to use for the Coordinate Reference System.

You may also specify tags to add to the created Geoscience objects.

Then we call `convert_adamf`, passing it the ADAMF file path, EPSG code, the service manager widget from above, a path we want the published objects to appear under and finally whether we want existing objects to be overwritten.

Then we loop over the results from the `convert_adamf` function, printing out each object that was published to Evo.


```python
import os
import pprint

from evo.data_converters.adamf.importer import convert_adamf

adamf_file = os.path.join(os.getcwd(), "data/input/WP_assay.csv")
epsg_code = 32650

tags = {"TagName": "Adam's Tag value"}

objects_metadata = convert_adamf(
    filepath=adamf_file,
    coordinate_reference_system=epsg_code,
    service_manager_widget=manager,
    tags=tags,
    upload_path="notebook",
    overwrite_existing_objects=False,
)

print()
print("These objects have now been published:")

for metadata in objects_metadata:
    pprint.pp(metadata, indent=4)
```
